```python
#OUTLIER DETECTION in CROP
#Function to calc lowerRange and UpperRange using 1.5IQR rule
def outlier(df1):
    #Sort in Ascending
    sorted(df1)
    #Q1 and Q3
    Q1,Q3 = np.nanpercentile(df1, [25,75])
    #Calc IQR
    IQR = Q3 - Q1
    #Calc lowerRange
    lowerRange = Q1 - (1.5 * IQR)
    #Calc upperRange
    upperRange = Q3 + (1.5 * IQR)
    #Return Ranges
    return lowerRange,upperRange
```

```python
from sklearn.utils import resample

# Separate Target Classes
df_1 = df1[df1.HeartDisease==0]

df_2 = df1[df1.HeartDisease==1]
 
# Upsample minority class
df_1_upsampled = resample(df_1, 
                                 replace=True,     # sample with replacement
                                 n_samples=498,    # to match majority class
                                 random_state=123) # reproducible results

# Combine majority class with upsampled minority class
df1 = pd.concat([df_2, df_1_upsampled])
 
# Display new class counts
df1.HeartDisease.value_counts()
```
